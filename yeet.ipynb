```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./data/train.csv')
```

```python
data.head()
```

```python
data.Embarked = data.Embarked.fillna('S')
```

```python
data.head()
```

```python
data.Cabin = data.Cabin.fillna("n/a")
```

```python
data.head()
```

```python
data.Fare.describe()
```

```python
data['Fare_Category'] = pd.cut(data['Fare'], bins = [-np.inf, 7.9104, 14.454200, 31, np.Inf], labels=['low', 'mid_low', 'mid_high', 'high'])
```

```python
data.head()
```

```python
data['Salutation'] = data.Name.apply(lambda name: name.split(',')[1].split('.')[0].strip())
```

```python
data.head()
```

```python
data.loc[data['Fare_Category'].isna()]
```

```python
train = data.drop(labels=['PassengerId', 'Survived', 'Name', 'Ticket', 'Fare', 'Cabin'], axis=1)
```

```python
train.head()
```

```python
grp = data.groupby(['Sex', 'Pclass'])
```

```python
grp.Age.apply(lambda x: x.fillna(x.median()))
data.Age.fillna(train.Age.median, inplace = True)
```

```python
train.head()
```

```python
data_sex = data[['Sex']]
data_fare = data[['Fare_Category']]
```

```python
from sklearn.preprocessing import OrdinalEncoder

sex_encoder = OrdinalEncoder()
data_sex_encoded = sex_encoder.fit_transform(data_sex) #male = 1, female = 0
fare_encoder = OrdinalEncoder(categories=[['low', 'mid_low', 'mid_high', 'high']])

data_fare_encoded = fare_encoder.fit_transform(data_fare)
```

```python
sex_encoder.categories_
```

```python
fare_encoder.categories_
```

```python
train['Sex'] = data_sex_encoded
```

```python
train.head()
```

```python
train['Fare_Category'] = data_fare_encoded
```

```python
train.head()
```

```python
from sklearn.preprocessing import LabelBinarizer

embarked_encoder = LabelBinarizer()
```

```python
data_embarked = data['Embarked']
```

```python
data_embarked_onehot = embarked_encoder.fit_transform(data_embarked)
```

```python
data_embarked_onehot.shape
```

```python
data_embarked_df = pd.DataFrame(data_embarked_onehot)
```

```python
data_embarked_df.head()
```

```python
data_embarked_df = data_embarked_df.rename(columns={0: "Embarked_C", 1: "Embarked_Q", 2: "Embarked_S"})
```

```python
train = pd.concat([train, data_embarked_df], axis=1).drop(['Embarked'], axis=1)
```

```python
train.head()
```

```python
salutations = pd.get_dummies(data[['Salutation']])
```

```python
train = pd.concat([train, salutations], axis=1).drop(['Salutation'], axis=1)
```

```python
train.head()
```
